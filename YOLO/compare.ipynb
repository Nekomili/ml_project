```python
import os
from ultralytics import YOLO
import torch
```

```python
def list_files(directory):
  try:
    filenames = os.listdir(directory)
    return filenames
  except FileNotFoundError:
    return []
```

```python
models = list_files("./paras_yolo_malli")
print(models)
```

```python
for node in models:
    print(node)
    model = YOLO("./paras_yolo_malli/" + node)
    model.val()
    input()
```
