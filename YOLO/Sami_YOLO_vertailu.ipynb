```python
import json
from ultralytics import YOLO
from pathlib import Path

# Polku dataset-konfigiin
data_config = 'config/materials.yaml'

# Mallit ja niiden nimet
models = {
    'YOLO11s': 'yolo11s.pt',
    'YOLO11l': 'yolo11l.pt',
    'YOLO8s': 'yolov8s.pt',
    'YOLO8l': 'yolov8l.pt'
}

# Tallennetaan metriikat tÃ¤nne
results_summary = {}

# Koulutus- ja validointisilmukka
for name, weights in models.items():
    print(f"\nðŸš€ Koulutetaan malli: {name}")
    model = YOLO(weights)

    # Koulutus oletusparametreilla (epochs=100, imgsz=640 jne.)
    model.train(data=data_config, epochs=50)

    # Ajetaan validointi koulutuksen jÃ¤lkeen
    results = model.val(data=data_config)

    # Poimitaan tulokset: mAP50, mAP50-95, precision, recall
    metrics = results.box
    results_summary[name] = {
        'mAP50': round(metrics.map50, 4),
        'mAP50-95': round(metrics.map, 4),
        'precision': round(metrics.p, 4),
        'recall': round(metrics.r, 4)
    }

# Tallenna tulokset tiedostoon
output_file = Path("mallivertailu_tulokset.json")
with open(output_file, "w") as f:
    json.dump(results_summary, f, indent=4)

# Tulosta yhteenveto
print("\nðŸ“Š Vertailun tulokset:")
print(f"{'Malli':<10} {'mAP50':>8} {'mAP50-95':>10} {'Prec.':>8} {'Recall':>8}")
print("-" * 50)
for model_name, metrics in results_summary.items():
    print(f"{model_name:<10} {metrics['mAP50']:>8} {metrics['mAP50-95']:>10} {metrics['precision']:>8} {metrics['recall']:>8}")
```

```python
# Koulutus- ja validointisilmukka
for name, weights in models.items():
    print(f"\nðŸš€ Koulutetaan malli: {name}")
    model = YOLO(weights)

    # Koulutus oletusparametreilla (epochs=100, imgsz=640 jne.)
    model.train(data=data_config, epochs=50)

    # Ajetaan validointi koulutuksen jÃ¤lkeen
    results = model.val(data=data_config)

    # Poimitaan tulokset: mAP50, mAP50-95, precision, recall
    metrics = results.box
    results_summary[name] = {
        'mAP50': round(metrics.map50, 4),
        'mAP50-95': round(metrics.map, 4),
        'precision': round(metrics.p.mean(), 4),
        'recall': round(metrics.r.mean(), 4)
    }

# Tallenna tulokset tiedostoon
output_file = Path("mallivertailu_tulokset.json")
with open(output_file, "w") as f:
    json.dump(results_summary, f, indent=4)

# Tulosta yhteenveto
print("\nðŸ“Š Vertailun tulokset:")
print(f"{'Malli':<10} {'mAP50':>8} {'mAP50-95':>10} {'Prec.':>8} {'Recall':>8}")
print("-" * 50)
for model_name, metrics in results_summary.items():
    print(f"{model_name:<10} {metrics['mAP50']:>8} {metrics['mAP50-95']:>10} {metrics['precision']:>8} {metrics['recall']:>8}")
```
