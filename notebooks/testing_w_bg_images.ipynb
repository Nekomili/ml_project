```python
import sqlite3
import pandas as pd
import os
```

```python
with sqlite3.connect("../data/database/annotations.sqlite") as conn:
    cursor = conn.cursor()

    cursor.execute("SELECT file FROM training")
    files_t = set([node[0] for node in cursor.fetchall()])
    cursor.execute("SELECT file FROM validation")
    files_v = set([node[0] for node in cursor.fetchall()])

    print(len(files_t))
    print(len(files_v))
    cursor.execute("SELECT DISTINCT file FROM silver")
    all_files = [node[0] for node in cursor.fetchall()]
    print(f"All file len: {len(all_files)}")
```

```python
print(all_files)
bg_files = [node for node in all_files if node not in files_t and node not in files_v]
print(len(bg_files))
```

```python
def write_training_files(bg_files:list[str], output_dir = "../YOLO/datasets/materials", img_path = "/home/jovyan/shared/dataset/"):
    train_image_dir = os.path.join(output_dir, "images", "train")
    if not os.path.exists(train_image_dir):
        os.makedirs(train_image_dir)
    for filename in bg_files:
        img_filename = filename.split('.')[0]+".jpg"
        os.symlink(os.path.join(img_path, img_filename), os.path.join(train_image_dir, img_filename))

write_training_files(bg_files)
```
