```python
import os

files = os.listdir("../data/example_images")

print(files)
files.sort()
print(files)
```

```python
def divide_list_into_segments(input_list, segment_size=300):
    """
    Divides a list into segments of a specified size.

    Args:
        input_list: The list to be divided.
        segment_size: The desired size of each segment (default: 300).

    Returns:
        A list of lists, where each inner list is a segment of the input list.
    """
    if not isinstance(input_list, list):
        raise TypeError("Input must be a list.")
    if not isinstance(segment_size, int) or segment_size <= 0:
        raise ValueError("Segment size must be a positive integer.")

    return [input_list[i:i + segment_size] for i in range(0, len(input_list), segment_size)]
```

```python
segments = divide_list_into_segments(files, segment_size=300)
segments

names = ["sami", "joonas","mika", "minna","reetta"]

len(segments)
len(segments[-1])
```

```python
divided = {names[i]:segments[i] for i in range(5)}
divided
```

```python
import csv
with open("../data/divided.csv", 'w', newline='', encoding='UTF-8') as csvfile:
    writer = csv.writer(csvfile, delimiter=';')
    writer.writerow(['Name', 'Image Filename'])

    for name, images in divided.items():
        for image in images:
            writer.writerow([name, image])
```
