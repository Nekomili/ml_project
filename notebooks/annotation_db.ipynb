```python
import os
import pandas as pd
import io

PATH = "../data/"

contents = os.listdir(PATH)
label_dirs = [item for item in contents if "labels_" in item]

print(label_dirs)
```

```python
path = PATH + label_dirs[3]
labels = os.listdir(path)
empties = []

print(len(labels))

name = path.split("_")[-1]
list_of_dicts = []

for i, label in enumerate(labels):
    # <object-class> <x-center> <y-center> <width> <height>
    with io.open(f"{path}/{label}", mode='r', encoding='utf-8') as file:
        lines = file.readlines()
        if lines:
            for line in lines:
                data = line.strip().split(" ")
                row = {
                    "name":name,
                    "file":label,
                    "class":int(data[0]),
                    "x-center":float(data[1]),
                    "y-center":float(data[2]),
                    "width":float(data[3]),
                    "height":float(data[4])
                }
                list_of_dicts.append(row)
        else:
            row = {
                "name":name,
                "file":label,
                "class":int(1),
                "x-center":float(0),
                "y-center":float(0),
                "width":float(0),
                "height":float(0)
            }
            list_of_dicts.append(row)
```

```python
def read_files(path:str) -> pd.DataFrame:
    name = path.split("_")[-1]
    labels = os.listdir(path)
    list_of_dicts = []

    for label in labels:
        # <object-class> <x-center> <y-center> <width> <height>
        with io.open(f"{path}/{label}", mode='r', encoding='utf-8') as file:
            lines = file.readlines()
            if lines:
                for line in lines:
                    data = line.strip().split(" ")
                    row = {
                        "name":name,
                        "file":label,
                        "class":int(data[0]),
                        "x-center":float(data[1]),
                        "y-center":float(data[2]),
                        "width":float(data[3]),
                        "height":float(data[4])
                    }
                    list_of_dicts.append(row)
            else:
                row = {
                    "name":name,
                    "file":label,
                    "class":int(1),
                    "x-center":float(0),
                    "y-center":float(0),
                    "width":float(0),
                    "height":float(0)
                }
                list_of_dicts.append(row)
    return pd.DataFrame(list_of_dicts)
```

```python
dataframes = []
for dir in label_dirs:
    path = PATH + dir
    dataframes.append(read_files(path=path))

all_data = pd.concat(dataframes)
```
